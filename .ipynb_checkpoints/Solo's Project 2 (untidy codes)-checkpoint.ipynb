```python
selector = SelectKBest(f_classif, k=28)
selector.fit(nominal.apply(le.fit_transform), df['SalePrice'])
selector.pvalues_
```

```python
p_value_nominal = f_classif(nominal.apply(le.fit_transform), df['SalePrice'])
```

```python
nominal_feature_selection = zip(list(nominal.columns), list(p_value_nominal[1]))
for i in nominal_feature_selection:
    print (i)
```

Based on the ANOVA tests above, based on a critical p-value of 0.05, we shall choose to exclude the following variables:
1. Id- see ******
2. PID - see ******
3. MS SubClass
4. Alley
5. Lot Config
6. Condition 1
7. Building Type
8. House Style
9. Roof Style
10. Exterior 1st
11. Exterior 2nd
12. Misc Feature
13. Mo Sold
14. Sale Type

_****** Eliminating these features even tho p-value<0.05 cause:_

_(a) will lead to too many variables after OHE, might confuse model;_

_(b) Even if model successfully trained, when deployed, it will still see new values which are unseen (might crash the model then)_

Basically, we can categorize most of the variables into those 7 main factors.

We shall proceed from here and group the relevant variables into the 7 factors and see how we can feature engineer & extract for each group.
