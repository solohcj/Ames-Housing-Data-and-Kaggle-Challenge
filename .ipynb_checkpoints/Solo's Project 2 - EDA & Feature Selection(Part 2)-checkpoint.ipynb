## EDA: Preliminary Feature Elimination through Filter Methods

### a. Prelim Feature Elimination for Nominal Data

```python
nominal.head(3)
```

```python
# Label encoding the nominal variables to perform ANOVA (using f_classf from sklearn)
# Assuming intervals = 1 for label encoding

le = LabelEncoder()
p_value_nominal = f_classif(nominal.apply(le.fit_transform), y_train)
```

```python
nominal_feature_selection = zip(list(nominal.columns), list(p_value_nominal[1]))
for i in nominal_feature_selection:
    print (i)
```

Based on the ANOVA tests above and a critical p-value of 0.05, we shall choose to exclude the following features:
1. PID - see ******
2. MS SubClass
3. Alley
4. Lot Config
5. Condition 1
6. Building Type
7. House Style
8. Exterior 1st
9. Exterior 2nd
10. Misc Feature
11. Mo Sold
12. Sale Type


_****** Eliminating these features even tho p-value<0.05 cause:_

_(a) will lead to too many variables after OHE, might confuse model;_

_(b) Even if model successfully trained, when deployed, it will likely still see new values which are unseen (might crash the model then)_

```python
nominal_features_to_drop = ['PID', 'MS SubClass', 'Alley', 'Lot Config', 'Condition 1', 'Bldg Type', 'House Style', 'Exterior 1st', 'Exterior 2nd', 'Misc Feature', 'Mo Sold', 'Sale Type']
len(nominal_features_to_drop)
```

```python
# Dropping the features and checking

nominal_remaining = nominal.copy()
nominal_remaining.drop(nominal_features_to_drop, axis=1, inplace=True)
nominal_remaining.head(3)
```

### b.  Prelim Feature Elimination for Numerical Data

```python
# Redefining numeric dataframe with the newly extracted/engineered features

numeric = X_train.drop(['Lot Shape', 'Utilities', 'Land Slope', 'Overall Qual', 'Overall Cond', 'Exter Qual', 'Exter Cond', 'Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2', 'Heating QC', 'Electrical', 'Kitchen Qual', 'Functional', 'Fireplace Qu', 'Garage Finish', 'Garage Qual', 'Garage Cond', 'Paved Drive', 'Pool QC', 'Fence', 'Bsmt Full Bath', 'Bsmt Half Bath', 'Full Bath', 'Half Bath', 'Bedroom AbvGr', 'Kitchen AbvGr', 'TotRms AbvGrd', 'Fireplaces', 'Garage Cars', 'PID', 'MS SubClass', 'MS Zoning', 'Street', 'Alley', 'Land Contour', 'Lot Config', 'Neighborhood', 'Condition 1', 'Condition 2', 'Bldg Type', 'House Style', 'Year Built', 'Year Remod/Add', 'Roof Style', 'Roof Matl', 'Exterior 1st', 'Exterior 2nd', 'Mas Vnr Type', 'Foundation', 'Heating', 'Central Air', 'Garage Type', 'Garage Yr Blt', 'Misc Feature', 'Yr Sold', 'Mo Sold', 'Sale Type'], axis=1)
```

```python
numeric.head(3)
```

```python
numeric.shape
```

```python
# Adding SalePrice variable to check for correlation
numeric['SalePrice'] = y_train
```

```python
numeric.describe()
```

```python
# For Numerical data we will use the Pearson's Correlation matrix
correlation_numeric = numeric.corr()
correlation_numeric
```

```python
# Heatmap to visualize things

plt.figure(figsize=(15,10))

mask = np.zeros_like(correlation_numeric, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(correlation_numeric, mask=mask);
```

```python
# We shall attempt to sieve out the features with correlation coefficient>|0.6| using a for loop

mask = correlation_numeric['SalePrice'] > abs(0.6)
corr_w_target = correlation_numeric[['SalePrice']]
corr_w_target[mask]
```

Based on the Pearson's tests above and a cut-off R value of |0.6|, we shall only choose the features in the dataframe above.

```python
numeric_features_to_drop = [i for i in numeric.columns if i not in list(corr_w_target[mask].index)]
len(numeric_features_to_drop)
```

```python
# Dropping the features and checking

numeric_remaining = numeric.copy()
numeric_remaining.drop(numeric_features_to_drop, axis=1, inplace=True)
numeric_remaining.head()
```

### c. Prelim Feature Elimination for Ordinal Data

**Our Strategy for Ordinal:** 

Assume as numerical data to do Pearson correlation, eliminate as appropriate, thereafter use as categorical (ordinal) data

_Because the order of the categories are not in proper sequence yet (ie. N.A is after the supposed largest category), we will need to encode our data manually based on the data documention :(_

```python
ordinal.head(3)
```

**The following features already have their data in numerical (ordinal) form and would not need to be transformed:**
1. Overall Qual
2. Overall Cond
3. Bsmt Half Bath
4. Bsmt Full Bath
5. Full Bath
6. Half Bath
7. Bedroom AbvGr
8. Kitchen AbvGr
9. TotRms AbvGr
10. Fireplaces
11. Garage Cars

_For the purpose of this project, we will **assume** that the **difference between the values of the ordinal features is 1**_

```python
ordinal['Lot Shape'] = ordinal['Lot Shape'].map({'IR3':0, 'IR2':1, 'IR1':2, 'Reg':3})
ordinal['Utilities'] = ordinal['Utilities'].map({'ELO':0, 'NoSeWa':1, 'NoSewr':2, 'AllPub':3})
ordinal['Land Slope'] = ordinal['Land Slope'].map({'Gtl':0, 'Mod':1, 'Sev':2})
ordinal['Exter Qual'] = ordinal['Exter Qual'].map({'Po':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
ordinal['Exter Cond'] = ordinal['Exter Cond'].map({'Po':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
ordinal['Bsmt Qual'] = ordinal['Bsmt Qual'].map({'No Basement':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
ordinal['Bsmt Cond'] = ordinal['Bsmt Cond'].map({'No Basement':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
ordinal['Bsmt Exposure'] = ordinal['Bsmt Exposure'].map({'No Basement':0, 'No':1, 'Mn':2, 'Av':3, 'Gd':4})
ordinal['BsmtFin Type 1'] = ordinal['BsmtFin Type 1'].map({'No Basement':0, 'Unf':1, 'LwQ':2, 'Rec':3, 'BLQ':4, 'ALQ':5, 'GLQ':6})
ordinal['BsmtFin Type 2'] = ordinal['BsmtFin Type 2'].map({'No Basement':0, 'Unf':1, 'LwQ':2, 'Rec':3, 'BLQ':4, 'ALQ':5, 'GLQ':6})
ordinal['Heating QC'] = ordinal['Heating QC'].map({'Po':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
ordinal['Electrical'] = ordinal['Electrical'].map({'Mix':0, 'FuseP':1, 'FuseF':2, 'FuseA':3, 'SBrkr':4})
ordinal['Kitchen Qual'] = ordinal['Kitchen Qual'].map({'Po':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
ordinal['Functional'] = ordinal['Functional'].map({'Sal':0, 'Sev':1, 'Maj2':2, 'Maj1':3, 'Mod':4, 'Min2':5, 'Min1':6, 'Typ':7})
ordinal['Fireplace Qu'] = ordinal['Fireplace Qu'].map({'No Fireplace':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
ordinal['Garage Finish'] = ordinal['Garage Finish'].map({'No Garage':0, 'Unf':1, 'RFn':2, 'Fin':3})
ordinal['Garage Qual'] = ordinal['Garage Qual'].map({'No Garage':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
ordinal['Garage Cond'] = ordinal['Garage Cond'].map({'No Garage':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
ordinal['Paved Drive'] = ordinal['Paved Drive'].map({'N':0, 'P':1, 'Y':2})
ordinal['Pool QC'] = ordinal['Pool QC'].map({'No Pool':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
ordinal['Fence'] = ordinal['Fence'].map({'No Fence':0, 'MnWw':1, 'GdWo':2, 'MnPrv':3, 'GdPrv':4})
```

```python
ordinal['SalePrice'] = y_train
ordinal.head()
```

```python
correlation_ordinal = ordinal.corr()
correlation_ordinal
```

```python
# Heatmap to visualize things

plt.figure(figsize=(15,10))

mask = np.zeros_like(correlation_ordinal, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(correlation_ordinal, mask=mask);
```

```python
# We shall attempt to sieve out the features with correlation coefficient>|0.4| using a for loop

mask = correlation_ordinal['SalePrice'] > abs(0.4)
corr_w_target_ordinal = correlation_ordinal[['SalePrice']]
corr_w_target_ordinal[mask]
```

Based on the Pearson's tests above and a cut-off R value of |0.4|, we shall the features shown above

```python
ordinal_features_to_drop = [i for i in ordinal.columns if i not in list(corr_w_target_ordinal[mask].index)]
ordinal_features_to_drop
```

```python
# Dropping the features and checking

ordinal_remaining = ordinal.copy()
ordinal_remaining.drop(ordinal_features_to_drop, axis=1, inplace=True)
ordinal_remaining.head(3)
```

## Encoding Categorical(Ordinal & Nominal) Features

```python
# Combining lists of features to be dropped & dropping them from combined dataframe

features_to_drop = nominal_features_to_drop + numeric_features_to_drop + ordinal_features_to_drop

X_train.drop(features_to_drop, axis=1, inplace=True)
X_train.head(3)
```

```python
X_train.shape
```

### a. Encoding Ordinal Data

```python
X_train['Exter Qual'] = X_train['Exter Qual'].map({'Po':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
X_train['Bsmt Qual'] = X_train['Bsmt Qual'].map({'No Basement':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
X_train['Kitchen Qual'] = X_train['Kitchen Qual'].map({'Po':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
X_train['Bsmt Exposure'] = X_train['Bsmt Exposure'].map({'No Basement':0, 'No':1, 'Mn':2, 'Av':3, 'Gd':4})
X_train['Heating QC'] = X_train['Heating QC'].map({'Po':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
X_train['Fireplace Qu'] = X_train['Fireplace Qu'].map({'No Fireplace':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
X_train['Garage Finish'] = X_train['Garage Finish'].map({'No Garage':0, 'Unf':1, 'RFn':2, 'Fin':3})
X_train.head()
```

### b. Encoding Ordinal Data

```python
X_encoded = pd.get_dummies(X_train, drop_first=True)
```

```python
X_encoded.head()
```

```python
X_encoded.shape
```
